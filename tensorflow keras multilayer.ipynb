```python
(xtrain,ytrain),(xtest,ytest)=mnist.load_data()
```

```python
xtrain=xtrain/255
xtest=xtest/255
```

```python
xflat=xtrain.reshape(len(xtrain),28*28)
xflattest=xtest.reshape(len(xtest),28*28)
model=keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,),activation="sigmoid")])
```

```python
model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
```

```python
model.fit(xflat,ytrain,epochs=5)  #hence accuracy is improved
```

```python
#lets  test the data
```

```python
model.evaluate(xflattest,ytest)   #testing accuracy on test data
```

```python
ypred=model.predict(xflattest)
```
